### Importing Library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import requests
```

### URL from IPL Auction 2023

```python
URL='https://www.iplt20.com/auction/2022'
```

```python
page = requests.get(URL)
```

```python
print(page.text)
```

### Soup containing all data

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

### Finding all the table anchor.

```python
table=soup.find('table',class_='ih-td-tab auction-tbl')
```

```python
table
```

### finding all th inside table

```python
headers = table.find_all('th')
```

```python
headers
```

### filtering all table columns

```python
headers = []
for th in table.find_all('th'):
    name = th.text.strip()
    headers.append(name)
```

```python
headers
```

### Adding columns to DataFrame

```python
df=pd.DataFrame(columns=headers)
```

```python
df
```

### finding all tr .i.e table rows

```python
rows=table.find_all('tr')
```

```python
for i in rows[1:]:
    data=i.find_all('td')
    row=[tr.text for tr in data]
    l=len(df)
    df.loc[l]=row
```

```python
df #adding all the rows to the dataframe
```

### Replacing the \n from Team 

```python
df['TEAM']=df['TEAM'].str.replace('\n','')
```
